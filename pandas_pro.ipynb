## EDA

```python
#Top_Ten_repetative_restaurant_type

Top_Ten_Types = df['restaurant type'].value_counts()[:10]
Top_Ten_Types
```

```python
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')
sns.barplot(x=Top_Ten_Types.values, y=Top_Ten_Types.index)
plt.title('Top Restaurant Types' , size = 20)
plt.xlabel('Number' , size = 15)
plt.ylabel('Kind' , size = 15)
plt.show()
```

```python
#list of restaurant by ratings


df[['restaurant name','cuisines type' ,'rate (out of 5)','num of ratings' ]].sort_values(by=['num of ratings'],ascending=False)
```

```python
print(df.columns)
```

```python
#Cost range of restaurant

cheap=df[['restaurant name','avg cost (two people)']].sort_values(by='avg cost (two people)' , ascending=True)[:10]
print(cheap)
```

```python
#Low Budget Restaurants 

plt.figure(figsize=(10, 6))
plt.barh(cheap['restaurant name'], cheap['avg cost (two people)'],color = "#0c6a96")
plt.xlabel('Average Cost (Two People)')
plt.ylabel('Restaurant Name')
plt.title('Low Budget Restaurants')
plt.gca().invert_yaxis()  # Invert y-axis to have the lowest cost at the top
plt.show()
```

```python
#restaurant with online options


df[['restaurant name','online_order','area']].sort_values(by='online_order' , ascending=False)[:10]
```

```python
print(df.columns)
```

```python
#Illustration of both options



online_order_counts = df['online_order'].value_counts()
table_booking_counts = df['table booking'].value_counts()

plt.figure(figsize=(10, 5))
colors = ['#06bf0c', '#6b0696'] 
plt.subplot(1, 2, 1)
plt.pie(online_order_counts, labels=online_order_counts.index, autopct='%1.1f%%', explode=[0.05, 0],colors=colors, startangle=140)
plt.title('Online Order options')
colors = ['#065fbf', '#b006bf']
# Plotting the second pie chart for 'table_booking options'
plt.subplot(1, 2, 2)
plt.pie(table_booking_counts, labels=table_booking_counts.index, autopct='%1.1f%%', explode=[0.05, 0],colors=colors, startangle=140)
plt.title('Table Pre-Booking options')
plt.tight_layout()
plt.show()
```
